```python
from pinecone import Pinecone
import os
from dotenv import load_dotenv
load_dotenv()
from langchain.chat_model import init_chat_model
import re
from pinecone_plugins.assistant.models.chat import Message
import tempfile

PINECONE_API_KEY = os.getenv('PINECONE_API_KEY')

pc = Pinecone(api_key=PINECONE_API_KEY)

assistant = pc.assistant.Assistant(
    assistant_name="paperai", 
)
```

```python
book = "About the book: In 'The Psychology of Money,' Morgan Housel reveals that successful financial management is less about mathematical formulas and more about human behavior. With 19 engaging stories, Housel delves into the intricate ways our emotions, personal histories, and social influences shape our financial decisions. He emphasizes that while traditional financial teachings focus on data, real-life choices often stem from complex motivations that unfold in everyday settings. This compelling exploration offers valuable insights into how we can improve our relationship with money and make wiser financial choices in a world where psychology trumps mathematics.About the author: Morgan Housel is a partner at The Collaborative Fund and a distinguished author known for his insights into finance and behavioral psychology. With a notable track record, he has received the Best in Business Award from the Society of American Business Editors and Writers on two occasions, and has been honored with the New York Times Sidney Award. He is also a two-time finalist for the prestigious Gerald Loeb Award for Distinguished Business and Financial Journalism. Residing in Seattle, Morgan shares his life with his wife and two children"
```

```python
data = []
record = []

splitted_book = re.split(r'[,.!?:]', book)
for individual in splitted_book:
    clean_book = individual.strip()
    data.append(clean_book)

for index, content in enumerate(data):
    record.append({"id":index, "content":content})
```

```python
with tempfile.NamedTemporaryFile(mode="w+", suffix=".txt", delete=False) as tmp_file:
    for line in book:
        tmp_file.write(line + "\n")
    tmp_file_path = tmp_file.name
tmp_file_path
```

```python
response = assistant.upload_file(
    file_path=tmp_file_path,
    timeout=None
)
```
